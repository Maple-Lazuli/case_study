```python
data['publish_time'] = pd.to_datetime(data['publish_time'], infer_datetime_format=True)
```

```python
data['trending_date'] = pd.to_datetime(data['trending_date'], format="%y.%d.%m")
```

```python
data = data.sort_values(by = 'trending_date', ascending = True)
```

```python
def prep_linear_regression(df):
    """
    Prepares the data for linear regressions. It assumes the data is sorted
    """
    data_dict = dict()
    keys = list(data['video_id'].unique())
    for key in keys:
        data_dict[key] = {'likes': [], 'dislikes': [], 'views': []}
    for row in range(len(df)):
        current_dict =  data_dict[df.iloc[row]['video_id']]
        current_dict['likes'].append(df.iloc[row]['likes'])
        current_dict['dislikes'].append(df.iloc[row]['dislikes'])
        current_dict['views'].append(df.iloc[row]['views'])
    return data_dict
        
    
    
    
```

```python
video_timeseries = prep_linear_regression(data)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
video_timeseries
```

```python
count = 0
linear_regressions_dict = []
for key in video_timeseries.keys():
    temp_dict = dict()
    likes = video_timeseries[key]['likes']
    dislikes = video_timeseries[key]['dislikes']
    views = video_timeseries[key]['views']
    if len(likes) > 3:
        length = len(likes)
        x = np.arange(length, dtype=float).reshape((length, 1))
        l =  LinearRegression().fit(x, np.array(likes).reshape(length,1))
        d =  LinearRegression().fit(x, np.array(dislikes).reshape(length,1))
        v =  LinearRegression().fit(x, np.array(views).reshape(length,1))
        linear_regressions_dict.append({'id': key, 'likes':l.coef_[0][0], 'dislikes':d.coef_[0][0] , 'views':v.coef_[0][0]})
    else:
        count = count + 1
        
```

```python
lr_df = pd.DataFrame.from_dict(linear_regressions_dict)
```
