```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("data/USvideos.csv")
def cleaner(string):
    string = string.lower()
    string = "".join([l for l in string if l not in '!\"#$%&\'()*+,-./:;<=>?@[\\]^_`{}~ ' ])
    lst = string.split("|")
    return set(lst)
df['tags_cleaned'] = df['tags'].apply(lambda x: cleaner(x))
def get_tag_counts(df):
    """
    Counts the number of occurrences in the list.
    RTN: Returns a dictionary
    """
    ## Get set of keys
    keys = set()
    for tag_set in df['tags_cleaned']:
        for tag in tag_set:
            keys.add(tag)
    
    ## Fill out the list
    dict_lst  = []
    for key in keys:
        standard_dict = {
        "tag_name":str(key),
        "views": 0, 
        "count": 0,
        'dislikes':0,
        'likes':0,
        'video_ids': set()
        }
        dict_lst.append(standard_dict)
        
    # Grab the data
    for row in range(len(df)):
        print(f'row {row} of {len(df)}')
        for tag in df.iloc[row]['tags_cleaned']:
            for element in dict_lst:
                if element['tag_name'] == tag:
                    element['views'] += df.iloc[row]['views']
                    element['count'] += 1
                    element['dislikes'] += df.iloc[row]['dislikes']
                    element['likes'] += df.iloc[row]['likes']
                    element['video_ids'].add(df.iloc[row]['video_id'])
                    
    return pd.DataFrame.from_dict(dict_lst)

tag_df = get_tag_counts(df)
```

```python
tag_df.to_json("tag_df_fix.json")
```
